```python
# Importação das bibliotecas
import pandas as pd    #Manipulação e análise de dados que fornece estruturas de dados flexíveis e permite trabalhar com dados estruturados ou heterogêneos.
import numpy as np    #Suporte a arrays e matrizes multidimensionais, possuindo uma larga coleção de funções matemáticas para trabalhar com estas estruturas.
import matplotlib.pyplot as plt    #Usado para controlar vários aspectos da criação de gráficos.
import seaborn as srn    #Visualização de dados baseada em Matplotlib que fornece uma interface de alto nível para desenhar gráficos estatísticos atraentes.
```

```python
#importar dados
tabela = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/dados/Dados.csv", sep=";")
#visulizar
tabela.head()
```

```python
srn.histplot(tabela.PIB, kde=True, bins=10).set(title='PIB')
```

```python
srn.histplot(tabela.VALOREMPENHO, kde=True, bins=10).set(title='Empenho')
```

```python
agrupado = tabela.sort_values('PIB').head(10)
agrupado = agrupado.iloc[:,1:3]
agrupado
agrupado.plot.bar(x='MUNICIPIO',y='PIB', color = 'gray')
```

```python
agrupado = tabela.sort_values('VALOREMPENHO').head(10)
agrupado = agrupado.iloc[:,[1,3]]
agrupado
agrupado.plot.bar(x='MUNICIPIO',y='VALOREMPENHO', color = 'gray')
```
